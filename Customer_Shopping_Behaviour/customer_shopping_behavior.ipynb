```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns = {'purchase_amount_(usd)': 'purchase_amount'})
```

```python
df.columns
```

```python
# Create a column age_group
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# Create column purchase_frequency_days

frequency_mapping = {
    'Fortnightly':14,
    'Monthly':30,
    'Weekly':7,
    'Quaterly':90,
    'Bi-Weekly': 14,
    'Yearly':365,
    'Every 3 months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
(df['purchase_amount'] / df['purchase_frequency_days']).all()
```

```python
df = df.drop('promo_code_used',axis=1)
```

```python
# Install first:
pip install sqlalchemy pymysql pandas

import pandas as pd
from sqlalchemy import create_engine, text
from sqlalchemy.engine import URL

# =====================================================
# MySQL Configuration
# =====================================================

MYSQL_USER = "root"
MYSQL_PASSWORD = "root@1998"
MYSQL_HOST = "127.0.0.1"
MYSQL_PORT = 3306
MYSQL_DATABASE = "cs"

TABLE_NAME = "mytable"

# =====================================================
# Create SQLAlchemy URL (safe for special characters)
# =====================================================

db_url = URL.create(
    drivername="mysql+pymysql",
    username=MYSQL_USER,
    password=MYSQL_PASSWORD,
    host=MYSQL_HOST,
    port=MYSQL_PORT,
    database=MYSQL_DATABASE
)

# =====================================================
# Connect to MySQL
# =====================================================

try:
    engine = create_engine(db_url)

    with engine.connect() as conn:
        print("✅ Connected to MySQL successfully")

except Exception as e:
    print("❌ Connection failed")
    print(e)
    raise

# =====================================================
# Example DataFrame
# Remove this section if df already exists
# =====================================================

try:
    df
    print("✅ Existing DataFrame found")
except NameError:
    print("⚠ df not found. Creating sample DataFrame.")

    df = pd.DataFrame({
        "id": [1, 2, 3],
        "name": ["John", "Mary", "David"],
        "salary": [50000, 60000, 70000]
    })

# =====================================================
# Write DataFrame to MySQL
# =====================================================

try:
    df.to_sql(
        name=TABLE_NAME,
        con=engine,
        if_exists="replace",   # replace / append
        index=False
    )

    print(f"✅ Data written to table '{TABLE_NAME}'")

except Exception as e:
    print("❌ Error writing DataFrame")
    print(e)

# =====================================================
# Read Data Back
# =====================================================

try:
    query = f"SELECT * FROM {TABLE_NAME} LIMIT 5"

    result_df = pd.read_sql(query, engine)

    print("\n✅ Sample Data From MySQL:")
    print(result_df)

except Exception as e:
    print("❌ Error reading data")
    print(e)
```
