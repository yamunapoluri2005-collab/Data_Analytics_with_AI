```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

Products = []
Prices = []
Descriptions = []
Reviews = []

headers = {"User-Agent": "Mozilla/5.0"}
for i in range(1, 10):
    url = f"https://www.flipkart.com/search?q=5g+mobile+under+50000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&page={i}"

    r = requests.get(url, headers=headers)
    soup = BeautifulSoup(r.text, "lxml")

    box = soup.find("div", class_="QSCKDh dLgFEE")

    names = box.find_all("div", class_="RG5Slk")
    for i in names:
        name = i.text
        Products.append(name)
    #print("Products: ",Products)

    price = box.find_all("div", class_="hZ3P6w DeU9vF")
    for i in price:
        pri = i.text
        Prices.append(pri)
    #print("Prices: ",Prices)

    Discription = box.find_all("div", class_="CMXw7N")
    for i in Discription:
        dis = i.text
        Descriptions.append(dis)
    #print(Descriptions)

    Review = box.find_all("div", class_="MKiFS6")
    for i in Review:
        rev = i.text
        Reviews.append(rev)
    #print("Reviews: ",len(Reviews))

df = pd.DataFrame({
    "Products Name": pd.Series(Products),
    "Prices": pd.Series(Prices),
    "Descriptions": pd.Series(Descriptions),
    "Reviews": pd.Series(Reviews),
})
print(df)
df.to_csv("/Users/saravananannamalai/Documents/Data analatic with AI/Flipkart_5G_Mobile_Under_50000.csv")
```
